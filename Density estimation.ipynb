```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
k=2000
x=np.random.rand(k).reshape(math.ceil(k/2),2)
plt.scatter(x[:,0],x[:,1],c='r')
plt.show()
```

```python
mean=np.mean(x,axis=0)
sd=(x-mean)**2
sd=np.mean(sd,axis=0)
sd=sd**(1/2)
sd
```

```python
x_new=np.exp(-(x-mean)**2/(2*sd**2))/(((2*np.pi)**(0.5))*sd)
```

```python
prob=np.zeros(math.ceil(k/2)).reshape(len(x),1)
for i in range(0,len(x)):
    prob[i]=x[i][0]*x[i][1]
```

```python
plt.plot(x[:,0],x_new[:,0],c='r',linestyle=' ',marker='x')
plt.plot(x[:,1],x_new[:,1],c='b',linestyle=' ',marker='o')
plt.show()
```

```python
probmean=np.mean(prob)
y=np.zeros(math.ceil(k/2)).reshape(len(x),1)
for i in range(0,len(x)):
    if prob[i]<probmean:
        y[i]=1  
```
