```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x=np.random.rand(20).reshape(10,2)
x
```

```python
cluster1=random.choice(x)
cluster2=random.choice(x)
plt.scatter(cluster1[0],cluster1[1],marker='^',c='r',s=150)
plt.scatter(cluster2[0],cluster2[1],marker='^',c='b',s=150)
plt.scatter(x[:,0],x[:,1],c='black')
plt.show()
```

```python
d1=np.zeros(10)
d2=np.zeros(10)
for k in range(0,5):
    for i in range(0,len(d1)):
        d1[i]=((x[i][0]-cluster1[0])**2+(x[i][1]-cluster1[1])**2)**(1/2)
        d2[i]=((x[i][0]-cluster2[0])**2+(x[i][1]-cluster2[1])**2)**(1/2)
    red=np.zeros(20).reshape(10,2)
    blue=np.zeros(20).reshape(10,2)
    redcount=0
    bluecount=0
    for i in range (0,len(d1)):
        if d1[i]<d2[i]:
            red[i]=x[i]
            redcount+=1
        else:
            blue[i]=x[i]
            bluecount+=1
    print (k)
    red=red[red!=0].reshape(redcount,2)
    blue=blue[blue!=0].reshape(bluecount,2)
    plt.plot(red[:,0],red[:,1],marker='o',linestyle=' ',c='r')
    plt.plot(blue[:,0],blue[:,1],marker='o',linestyle=' ',c='b')
    plt.scatter(cluster1[0],cluster1[1],marker='^',c='r',s=150)
    plt.scatter(cluster2[0],cluster2[1],marker='^',c='b',s=150)
    plt.show()
    cluster1[0]=np.mean(red[:,0])
    cluster1[1]=np.mean(red[:,1])
    cluster2[0]=np.mean(blue[:,0])
    cluster2[1]=np.mean(blue[:,1])
    plt.plot(red[:,0],red[:,1],marker='o',linestyle=' ',c='r')
    plt.plot(blue[:,0],blue[:,1],marker='o',linestyle=' ',c='b')
    plt.scatter(cluster1[0],cluster1[1],marker='^',c='r',s=150)
    plt.scatter(cluster2[0],cluster2[1],marker='^',c='b',s=150)
    plt.title('After mean')
    plt.show()
```
